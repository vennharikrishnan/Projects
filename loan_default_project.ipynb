```python
dfe=df.drop(['ID','Accompany_Client','Client_Education','Client_Marital_Status','Client_Housing_Type','Client_Gender','Age_Days','Registration_Days','ID_Days','Own_House_Age','Mobile_Tag','Homephone_Tag','Workphone_Working','Application_Process_Day','Application_Process_Hour','Client_Permanent_Match_Tag','Client_Contact_Work_Tag','Score_Source_1','Social_Circle_Default','Type_Organization'],axis=1)
```

```python
dfe
```

```python
dfe.dtypes
```

```python
dfe['Client_Income']=dfe['Client_Income'].astype('str')
dfe['Client_Income']=dfe['Client_Income'].str.replace('$','0',regex=False)
dfe['Client_Income']=dfe['Client_Income'].astype('float')
dfe['Credit_Amount']=dfe['Credit_Amount'].astype('str')
dfe['Credit_Amount']=dfe['Credit_Amount'].str.replace('$','0',regex=False)
dfe['Credit_Amount']=dfe['Credit_Amount'].astype('float')
dfe['Loan_Annuity']=dfe['Loan_Annuity'].astype('str')
dfe['Loan_Annuity']=dfe['Loan_Annuity'].str.replace('$','0',regex=False)
dfe['Loan_Annuity']=dfe['Loan_Annuity'].str.replace('#VALUE!','0',regex=False)
dfe['Loan_Annuity']=dfe['Loan_Annuity'].astype('float')
```

```python
dfe['Population_Region_Relative']=dfe['Population_Region_Relative'].astype('str')
dfe['Population_Region_Relative']=dfe['Population_Region_Relative'].str.replace('@','0',regex=False)
dfe['Population_Region_Relative']=dfe['Population_Region_Relative'].str.replace('#','0',regex=False)
dfe['Population_Region_Relative']=dfe['Population_Region_Relative'].astype('float')
```

```python
dfe['Employed_Days']=dfe['Employed_Days'].astype('str')
dfe['Employed_Days']=dfe['Employed_Days'].str.replace('x','0',regex=False)
dfe['Employed_Days']=dfe['Employed_Days'].astype('float')
```

```python
dfe['Score_Source_3']=dfe['Score_Source_3'].astype('str')
dfe['Score_Source_3']=dfe['Score_Source_3'].replace('&','0',regex=False)
dfe['Score_Source_3']=dfe['Score_Source_3'].replace('nan','0',regex=False) 
dfe['Score_Source_3']=dfe['Score_Source_3'].astype('float')
```

```python
dfe.dtypes
```

```python
#changing the null value into NaN
import numpy as np
dfe.loc[dfe.Client_Income==0,'Client_Income']=np.NaN
dfe.loc[dfe.Credit_Amount==0,'Credit_Amount']=np.NaN
dfe.loc[dfe.Loan_Annuity==0,'Loan_Annuity']=np.NaN
dfe.loc[dfe.Employed_Days==0,'Employed_Days']=np.NaN
```

```python
dfe.isnull().sum()
```

```python
#filling the missing values
dfe['Client_Income']=dfe['Client_Income'].fillna(dfe['Client_Income'].mean())
dfe['Car_Owned']=dfe['Car_Owned'].fillna(dfe['Car_Owned'].mode()[0])
dfe['Bike_Owned']=dfe['Bike_Owned'].fillna(dfe['Bike_Owned'].mode()[0])
dfe['Active_Loan']=dfe['Active_Loan'].fillna(dfe['Active_Loan'].mode()[0])
dfe['House_Own']=dfe['House_Own'].fillna(dfe['House_Own'].mode()[0])
dfe['Child_Count']=dfe['Child_Count'].fillna(dfe['Child_Count'].mode()[0])
dfe['Credit_Amount']=dfe['Credit_Amount'].fillna(dfe['Credit_Amount'].mean())
dfe['Loan_Annuity']=dfe['Loan_Annuity'].fillna(dfe['Loan_Annuity'].mean())
dfe['Population_Region_Relative']=dfe['Population_Region_Relative'].fillna(dfe['Population_Region_Relative'].mean())
dfe['Employed_Days']=dfe['Employed_Days'].fillna(dfe['Employed_Days'].mean())
dfe['Client_Family_Members']=dfe['Client_Family_Members'].fillna(dfe['Client_Family_Members'].mode()[0])
dfe['Cleint_City_Rating']=dfe['Cleint_City_Rating'].fillna(dfe['Cleint_City_Rating'].mean())
dfe['Score_Source_2']=dfe['Score_Source_2'].fillna(dfe['Score_Source_2'].mean())
dfe['Phone_Change']=dfe['Phone_Change'].fillna(dfe['Phone_Change'].mean())
dfe['Credit_Bureau']=dfe['Credit_Bureau'].fillna(dfe['Credit_Bureau'].mean())
```

```python
dfe.isna().sum()
```

```python
#Encoding - Get Dummies
dummy=pd.get_dummies(dfe[['Client_Income_Type','Loan_Contract_Type','Client_Occupation']],drop_first=True)
```

```python
#joining
fd=pd.concat([dfe,dummy],axis=1)
```

```python
fd=fd.drop(['Client_Income_Type','Loan_Contract_Type','Client_Occupation'],axis=1)
fd
```

```python
x=fd.drop(['Default'],axis=1)
y=fd['Default']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
scale.fit(x_train)
x_train = scale.fit_transform(x_train)
x_test = scale.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
print("Accuracy Score = ",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_pred,y_test)
```
