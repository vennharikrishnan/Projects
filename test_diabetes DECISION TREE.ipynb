```python
#function creating to remove string N,Y,0,1 into two class int 0 and 1
def label(row):
    if isinstance(row['Outcome'],str) and 'N' in row['Outcome']:
        return 0
    if isinstance(row['Outcome'],str) and 'Y' in row['Outcome']:
        return 1
    if isinstance(row['Outcome'],str) and '0' in row['Outcome']:
        return 0
    if isinstance(row['Outcome'],str) and '1' in row['Outcome']:
        return 1
```

```python
data['Outcome'] = data.apply(label,axis=1) # transferring label into outcome
data.head()
```
